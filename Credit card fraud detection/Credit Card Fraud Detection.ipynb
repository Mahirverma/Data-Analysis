<h1>Imports</h1>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
import pickle
```

<h1>Loading the dataset</h1>

```python
df=pd.read_csv('creditcard.csv')
df.head()
```

<h1>Preprocessing and visualizing the data</h1>

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["Amount"]= StandardScaler().fit_transform(df["Amount"].values.reshape(-1, 1))
df["Time"]= StandardScaler().fit_transform(df["Time"].values.reshape(-1, 1))
```

```python
df.describe()
```

```python
df.hist(bins=30, figsize=(50,50))
```

```python
df.duplicated().any()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().any()
```

```python
df['Class'].value_counts()
```

<h1>Handling imbalance data</h1>

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
smote=SMOTE(random_state=42)
x_smote,y_smote=smote.fit_resample(x,y)
```

```python
y_smote.value_counts()
```

<h1>Splitting the dataset into train data and test data</h1>

```python
x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote,test_size=0.3,random_state=42)
```

<h1>Training the data into the model</h1>

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train,y_train)
```

<h1>Predicting on test data</h1>

```python
predictions=logistic_model.predict(x_test)
```

```python
mapping={0:'Genuine',1:'Fraud'}
result=list(map(mapping.get,predictions))
result
```

```python
print(accuracy_score(predictions,y_test))
```

<h1>Checking with another model</h1>

```python
sgd_model=SGDClassifier()
sgd_model.fit(x_train,y_train)
```

```python
predictions=sgd_model.predict(x_test)
```

```python
mapping={0:'Genuine',1:'Fraud'}
result=list(map(mapping.get,predictions))
result
```

```python
print(accuracy_score(predictions,y_test))
```

<h1>Saving model with better accuracy</h1>

```python
pickle.dump(logistic_model,open('Logistic_model.pkl','wb'))
```
